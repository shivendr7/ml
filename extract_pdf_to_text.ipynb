```python
import PyPDF2
```

```python

def getPDFFileContentToTxt(pdfFile):
  myPDFFile = PyPDF2.PdfFileReader(pdfFile)

  with open('sample_data/Head First Java by Kathy Sierra, Bert Bates (z-lib.org).txt', 'w') as pdf_output:
    for page in range(myPDFFile.getNumPages()):
      data = myPDFFile.getPage(page).extractText()
      pdf_output.write(data)
    
  with open('sample_data/Head First Java by Kathy Sierra, Bert Bates (z-lib.org).txt', 'r') as myPDFContent:
    return myPDFContent.read().replace('\n', ' ')
```

```python
pdfFileContent = getPDFFileContentToTxt('drive/MyDrive/design/Head First Java by Kathy Sierra, Bert Bates (z-lib.org).pdf')
```
