```python
map_dataset=listdir('/root/.keras/datasets')
print('map_dataset:', map_dataset)
inside_maps=listdir('/root/.keras/datasets/maps')
print('inside_maps:', inside_maps)
inside_train=listdir('/root/.keras/datasets/maps/train')
print('inside_train:', inside_train)
```
