```
import os
import pandas as pd
import numpy as np

df=pd.read_csv("https://data.heatonresearch.com/data/t81-558/auto-mpg.csv",na_values=['NA','?'])

pd.set_option('display.max_columns',7)
pd.set_option('display.max_rows',5)
display(df)
#map with dataframes
df['origin_name']=df['origin'].map({
    1:'North America',2:'Europe',3:'Asia'
})
df=df.reindex(np.random.permutation(df.index))
display(df)
```

```
#apply with dataframes--execute the provided function against each col and rows of dataframes
display(df[['displacement','horsepower']][:10])
efficiency=df.apply(lambda x: x['displacement']/x['horsepower'],axis=1)
display(efficiency[:10])
df['efficiency']=efficiency
display(df[:10])
```

Feature engineering with apply map and grouping

```
import pandas as pd
df=pd.read_csv('https://www.irs.gov/pub/irs-soi/16zpallagi.csv')
df=df.loc[(df['zipcode']!=0) & (df['zipcode']!=99999),['STATE','zipcode','agi_stub','N1']]
display(df)
```

```
df1=df.loc[df['agi_stub']!=1,['agi_stub']]
display(df1.count())
```

```
medians={1:12500,2:37500,3:62500,4:87500,5:112500,6:212500}
df['agi_stub']=df.agi_stub.map(medians)

display(df['agi_stub'])
```

```
groupsofzipcode=df.groupby(by='zipcode')
display(pd.DataFrame(groupsofzipcode))
df=pd.DataFrame(groupsofzipcode.apply(lambda x:sum(x['N1']*x['agi_stub'])/sum(x['N1'])))
#pd.set_option('display.max_columns', 0)
#pd.set_option('display.max_rows', 10)

display(df)
```
