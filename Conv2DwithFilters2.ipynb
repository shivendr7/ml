```
import numpy as np
import requests
from PIL import Image, ImageFile
from io import BytesIO
```

```
url="https://gh.cdn.sewest.net/assets/ident//news/d81c6286/Avenger_Blog_IronMan_Artwork_3-663lb99uz.jpg?quality=65&width=1280px&height=720px"
response= requests.get(url)
img=Image.open(BytesIO(response.content))
```

```
bAw=img.convert("L") #blackAndwhite
np.asarray(bAw).shape
```

```
from scipy import signal
import matplotlib.pyplot as plt
```

```
Image.fromarray(np.asarray(bAw))
```

```
filter=np.array([
                 [1,2,3,2,1],[2,4,6,4,2],[3,9,12,9,3],[2,4,6,4,2],[1,2,3,2,1] #Gaussian Filter
])
out=signal.convolve2d(bAw,filter,mode='same',boundary='symm')  #bAw is image not array !!
```

```
fig,aux=plt.subplots(figsize=(20,20))
aux.imshow(out,cmap='gray')
```
